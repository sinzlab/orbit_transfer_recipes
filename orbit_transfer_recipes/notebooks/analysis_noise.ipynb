## RDM for single image and varying noise

```python
def centering(K):
    n = K.shape[0]
    unit = np.ones([n, n])
    I = np.eye(n)
    H = I - unit / n

    return np.dot(np.dot(H, K), H)  # HKH are the same with KH, KH is the first centering, H(KH) do the second time, results are the sme with one time centering
    # return np.dot(H, K)  # KH


def rbf(X, sigma=None):
    GX = np.dot(X, X.T)
    KX = np.diag(GX) - GX + (np.diag(GX) - GX).T
    if sigma is None:
        mdist = np.median(KX[KX != 0])
        sigma = math.sqrt(mdist)
    KX *= - 0.5 / (sigma * sigma)
    KX = np.exp(KX)
    return KX


def kernel_HSIC(X, Y, sigma):
    return np.sum(centering(rbf(X, sigma)) * centering(rbf(Y, sigma)))


def linear_HSIC(X, Y):
    L_X = np.dot(X, X.T)
    L_Y = np.dot(Y, Y.T)
    return np.sum(centering(L_X) * centering(L_Y))


def linear_CKA(X, Y):
    hsic = linear_HSIC(X, Y)
    var1 = np.sqrt(linear_HSIC(X, X))
    var2 = np.sqrt(linear_HSIC(Y, Y))

    return hsic / (var1 * var2)


def kernel_CKA(X, Y, sigma=None):
    hsic = kernel_HSIC(X, Y, sigma)
    var1 = np.sqrt(kernel_HSIC(X, X, sigma))
    var2 = np.sqrt(kernel_HSIC(Y, Y, sigma))

    return hsic / (var1 * var2)

def RDM(X, dist_measure="corr"):
    X = (X - X.mean(dim=-1).unsqueeze(-1))
    if dist_measure == "corr":
        result = (X @ torch.transpose(X, 0, 1)) / \
                                 torch.ger(torch.norm(X, 2, dim=1), torch.norm(X, 2,dim=1))  
    elif dist_measure == "l2":
        mtx1 = X
        m = mtx1.size(0)
        p = mtx1.size(1)
        mmtx1 = torch.stack([mtx1]*m)
        mmtx2 = torch.stack([mtx1]*m).transpose(0, 1)
        result = torch.sum((mmtx1 - mmtx2)**2, 2).squeeze()
    return result

def RDM_comparison(X,Y, dist_measure="corr"):
    RDM_X = RDM(X, dist_measure).flatten().numpy()
    RDM_Y = RDM(Y, dist_measure).flatten().numpy()
    result = RDM_X @ RDM_Y.T
    result /= (X.shape[0]-1)**2
    return result

def similarity(X, Y, dist_measure="CKA"):
    if dist_measure=="CKA":
        return linear_CKA(X.numpy(),Y.numpy())
    else:
        return RDM_comparison(X,Y, dist_measure)
```

```python
num_samples = 3
noise_stds = np.arange(0,0.5,0.01)
images = []
for batch_idx, (inputs, targets) in enumerate(analyser.sample_loader):
    inputs, targets = inputs.to(analyser.device,dtype=torch.float), targets.to(analyser.device)
    for i, noise_std in enumerate(noise_stds):
        dpi = 150 
        sns.set()
        sns.set_style("whitegrid")
        sns.set_context("paper")
        fig, axs = plt.subplots(
            1, num_samples * 2, sharex=False, sharey=False,
            dpi=dpi, figsize=(num_samples*3,4)
            
        )
        fig.tight_layout()  # Or equivalently,  "plt.tight_layout()"
        # apply noise
        trainer = copy.deepcopy(analyser.experiment.trainer)
        trainer.noise_std = {noise_std: 1.0}
        module = NoiseAugmentation(config=trainer, device=analyser.device, data_loader=analyser.sample_loader,
                      seed=42)
        model, inputs_ = module.pre_forward(analyser.model, inputs, {}, train_mode=False)
        # Forward
        outputs = model(inputs_)
        inputs_ = inputs_.transpose(1, 3)
        #Accuracy:
        _, predicted = outputs["logits"].max(1)
        total = targets.size(0)
        correct = predicted.eq(targets).sum().item()
        acc = 100. * correct / total
        #plotting
        output = outputs["conv_rep"].view(-1,2048,1).detach().cpu().numpy()
        for sample_idx in range(num_samples):
            axs[2*sample_idx].imshow(inputs_[sample_idx, :].cpu().numpy(),  interpolation='nearest')
            axs[2*sample_idx].axis('off')
            axs[2*sample_idx].set_title("std={:01.2f}".format(noise_std))
            sns.heatmap(output[sample_idx],ax=axs[2*sample_idx+1], vmin=0.0, vmax=1.0, center=0.2)
        # Used to return the plot as an image rray
        st = fig.suptitle(name + " Accuracy: {:02.1f}".format(acc), fontsize=12)
        st.set_y(0.95)
        fig.subplots_adjust(top=0.85,right=0.9)
        fig.canvas.draw()       # draw the canvas, cache the renderer
        image = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
        image = image.reshape(fig.canvas.get_width_height()[::-1] + (3,))
        images.append(image)
        plt.close(fig)
    break
imageio.mimsave('./{}.gif'.format(name), images, format='GIF', duration=0.3)
```
